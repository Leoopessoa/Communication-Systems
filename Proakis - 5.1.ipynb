The received signal $r (t) = s (t) + n (t)$ in a communication system is passed
through an ideal LPF with bandwidth $W$ and unity gain. The signal component
$s(t)$ has a power-spectral density

$$ S_s(f) = \frac{P_0}{1+(f/B)^2}$$

where $B$ is the 3-dB bandwidth. The noise component $n(t)$ has a power-spectral
density $N_0/2$ for all frequencies. Determine and plot the SNR as a function of the
ratio $W/B$. What is the filter bandwidth $W$ that yields a maximum SNR?

With the power-spectral density, we can have the noise power as:

$$P_n = \int^\infty _ {-\infty}S_n(f)df = \int^W _ {-W}S_n(f)df = N_0W$$

Now, for the signal power, we have:

$$P_s = \int^W _ {-W}S_s(f)df = \int^W _ {-W} \frac{P_0}{1+(f/B)^2}$$

With trigonometrical identities, we have:

$$P_s = P_0B \arctan \left(\frac W B\right)\bigg| ^W _{-W} = 2P_oB\arctan\left(\frac W B\right)$$

Thus, we have for the SNR:

$$SNR = \frac{P_s}{P_n} = \frac{2P_oB\arctan\left(\frac W B\right)}{N_0W}$$

As the plot should be with $W/B$ ratio:

 $$SNR = \frac{2P_oB}{N_0}× \frac{\arctan\left(\frac W B\right)}{(W/B)}$$


```python
 # Imports

import matplotlib.pyplot as plt 
import numpy as np 

# With the function normalized for 2P0/N0 (2P0/N0 = 1)

t = np.linspace(0,20,1000) # With t = (W/B)

SNR = (np.arctan(t))/(t) # defining SNR with 2P0/N0 = 1 and (W/B) from 0 to 20

figure1 = plt.figure(figsize=[8,6])

plt.plot(t,SNR) # Plotting

plt.xlim(0,20) # Defining Limits and scale
plt.ylim(0,1)
plt.yticks(np.arange(0,1.1,0.1))

plt.xlabel('W/B', fontsize = 12) # Labels 
plt.ylabel('SNR', fontsize = 12)
plt.title('SNR as a Function of (W/B) with 2P0/N0 = 1', fontsize = 16)

plt.grid()


```
